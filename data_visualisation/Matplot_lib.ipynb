**Import the matplot Library**


```python
import matplotlib.pyplot as plt
```

**Basic Plotting**

```python
x = [1, 2, 3, 4, 5, 6.2]
y = [10, 20, 25, 30, 50, 0]

plt.plot(x, y)
plt.show
```

```python
x = [1, 2, 3, 4, 5, 6.2]
y = [10, 20, 25, 30, 50, 0]

plt.plot(x, y)

#labeling

plt.xlabel("X_Label")
plt.ylabel("Y_Label")
plt.title("Line Graph")
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6.2]
y = [10, 20, 25, 30, 50, 0]

plt.plot(x, y, linestyle="--", color="Red", marker= "o" )
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6.2]
y = [10, 20, 25, 30, 50, 0]


plt.plot(x, y)


plt.grid(True)
plt.legend(["Data Line"])
plt.xlim(0, 7)
plt.ylim(0, 100)

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6.2]
y = [10, 20, 25, 30, 50, 0]


plt.scatter(x, y, marker="s", color="Red")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(categories, values, color = "green")
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, color="red", edgecolor="Black", bins=5)  #histogram plot
#plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")

```

```python
plt.subplot(1, 2, 1)  # (rows, columns, index)
plt.plot(x, y, 'r')
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(categories, values)
plt.title("Plot 2")

plt.tight_layout()
```
