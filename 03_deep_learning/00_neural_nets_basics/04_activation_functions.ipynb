# **Why Activation Functions are Needed**


*   Introduce non-linearity into neural networks.
*   Without them, even deep networks behave like linear models.
*   Allow networks to approximate complex patterns.

# **Common Activation Functions**

| Function                 | Formula                                        | Notes                                                     |
| ------------------------ | ---------------------------------------------- | --------------------------------------------------------- |
| **Sigmoid**              | $\sigma(x) = \frac{1}{1 + e^{-x}}$             | Smooth, but suffers from vanishing gradient.              |
| **Tanh**                 | $\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$ | Centered at 0, but still suffers from vanishing gradient. |
| **ReLU**                 | $f(x) = \max(0, x)$                            | Fast, simple, most commonly used.                         |
| **Leaky ReLU**           | $f(x) = \max(\alpha x, x)$                     | Fixes ReLU dying neuron problem.                          |
| **ELU**, **Swish**, etc. | Various                                        | Advanced, modern variants.                                |


# **Visualize Each Activation Function**

Use matplotlib to plot input vs. output for each function. Example:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

def relu(x): return np.maximum(0, x)
def sigmoid(x): return 1 / (1 + np.exp(-x))
def tanh(x): return np.tanh(x)

plt.plot(x, relu(x), label='ReLU')
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.legend()
plt.title("Activation Functions")
plt.grid(True)
plt.show()

```
