# **1. What is a Neuron?**

A neuron takes inputs, multiplies them with weights, adds bias, and passes the result through an activation function.

# **Single Neuron**

```python
import numpy as np

# Input features (x1, x2, x3)
X = np.array([0.5, 0.3, 0.2])

# Weights for the inputs
W = np.array([0.4, 0.7, 0.1])

# Bias
b = 0.2

# Linear combination (z = w1*x1 + w2*x2 + w3*x3 + b)
z = np.dot(W, X) + b
print("Weighted sum (z):", z)
```

# **Activation Functions**


*   Sigmoid  
*   ReLU
*   Tanh












**Sigmoid**

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

a = sigmoid(z)
print("Sigmoid activation (a):", a)
```

**ReLU**

```python
def relu(z):
    return np.maximum(0, z)

print("ReLU activation:", relu(z))
```

**Tanh**

```python
def tanh(z):
    return np.tanh(z)

print("Tanh activation:", tanh(z))
```

# **Wrap as a Neuron Class**

```python
class Neuron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def forward(self, x):
        z = np.dot(self.weights, x) + self.bias
        return self.sigmoid(z)

neuron = Neuron([0.4, 0.7, 0.1], 0.2)
output = neuron.forward([0.5, 0.3, 0.2])
print("Neuron output:", output)
```
